```python
import random

def get_word():
    words_list = ['Variable', 'List', 'Tuple', 'Dictionary', 'Function', 'Class', 'Inheritance',
                  'Module', 'Package', 'Loop', 'Conditional', 'Exception', 'Lambda', 'Decorator',
                  'Generator', 'Slice', 'Recursion', 'API', 'Framework', 'IDE']
    word = random.choice(words_list)
    return word.lower()

def display_word(word, guessed_letters):
    displayed = ""
    for letter in word:
        if letter in guessed_letters:
            displayed += letter
        else:
            displayed += "-"
    return displayed

def hangman():
    max_attempts = 6
    guessed_letters = []
    guessed_words = []
    word_to_guess = get_word()
    attempts = 0

    print("Welcome to Hangman!")
    print("The catagory is Python")
    print("You have 6 attempts to guess the word.")

    while True:
        print("\n" + display_word(word_to_guess, guessed_letters))

        guess = input("Guess a letter or the whole word: ").lower()

        if guess in guessed_letters:
            print("You already guessed that letter. Try again.")
            continue

        guessed_letters.append(guess)

        if len(guess) == 1 and guess.isalpha():
            if guess not in word_to_guess:
                attempts += 1
                print("Incorrect guess! Attempts remaining:", max_attempts - attempts)
        elif len(guess) == len(word_to_guess) and guess.isalpha():
            if guess in guessed_words:
                print("You've already guessed that word")
            elif guess != word_to_guess:
                print(f"{guess} is not the word")
                attempts += 1
                guessed_words.append(guess)
            else:
                print("Congratulations! You guessed the word:", word_to_guess)
                break
        else:
            print("Invalid input. Please enter a valid letter or the entire word.")

        if attempts >= max_attempts:
            print("Sorry, you ran out of attempts. The word was:", word_to_guess)
            break

if __name__ == "__main__":
    hangman()
```
